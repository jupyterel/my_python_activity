```python
# 3.1.9 Practical Activity
# How many films from each year (released from 2012 to the present) were watched on Netflix?

mo_gpby = df_1.groupby('Year')[['Netflix']].agg('sum').reset_index()

mo_gpby[mo_gpby['Year']>=2012]
```

```python
# What is the average runtime of movies released each year?
# My solution

# mo_gpby1_mine = df_1.groupby('Runtime')[['Netflix', 'Hulu', 'Prime Video', 'Disney+']].sum().agg('mean').reset_index()
# mo_gpby1_mine
```

```python
# Course solution
mo_gpby1 = df_1.groupby(['Year'])['Runtime'].agg('mean').reset_index()
mo_gpby1[mo_gpby1['Year'] >= 2012]
```

```python
# What are the best and worst reviews movies received on Rotten Tomatoes?
# My solution
mo_gpby_2_mine = df_1.groupby('Year')[['Rotten Tomatoes']].agg('sum').reset_index() \
.sort_values("Rotten Tomatoes", ascending=False)

mo_gpby_2_mine
```

```python
mo_gpby_2 = df_1.groupby(['Year'])['Rotten Tomatoes'].agg(['min','max']).reset_index()

mo_gpby_2[mo_gpby_2['Year']>=2012]
```
