```python
# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import datetime
```

```python
transport = pd.read_excel('transport_costs.xlsx')

transport
```

```python
# View the DataFrame.
print(transport.shape)
print(transport.columns)
print(transport.dtypes)
transport.head()
```

```python
new_df = transport[['region', 'country', 'port', 'sea freight cost',
       'road transport cost per km']]

new_df
```

```python
# Create new columns with calculated values.
new_df['total_sea_freight_costs'] = new_df['sea freight cost'] * 2 * 4

# View output.
new_df
```

```python
new_df['total_road_transport_costs'] = new_df['road transport cost per km'] * 10000

new_df
```

```python
new_df['total_export_costs'] = new_df['total_sea_freight_costs'] + new_df['total_road_transport_costs']

new_df
```

```python
# Export csv file.

new_df.to_csv('transport_python.csv')
```
